```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("unrate.csv")
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.head(12)
```

```python
x_values = df["DATE"].head(12)
y_values = df["VALUE"].head(12)
```

```python
plt.plot(x_values, y_values)
plt.show()
```

```python
plt.plot(x_values, y_values, color='r')
plt.xticks(rotation = 90)
```

```python
plt.plot(x_values, y_values, color='g')
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.plot(df['DATE'].head(12),df['VALUE'].head(12))
ax2.plot(df.DATE.loc[12:24],df.VALUE.loc[12:24])
```

```python
fig = plt.figure(figsize=(12,5))
for i in range(2):
    ax = fig.add_subplot(2,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = df[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])
```

```python
fig = plt.figure(figsize = (12,12))
for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = df[start_index:end_index]
    ax.plot(subset["DATE"], subset["VALUE"])
plt.show()
```

```python
df['MONTH'] = df['DATE'].dt.month
fig = plt.figure(figsize=(6,3))
plt.plot(df[0:12]['MONTH'], df[0:12]['VALUE'], c='red', label='1948')
plt.plot(df[12:24]['MONTH'], df[12:24]['VALUE'], c='blue', label='1949')
plt.legend()
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(df[0:12]['MONTH'], df[0:12]['VALUE'], c='red', label='1948')
plt.plot(df[12:24]['MONTH'], df[12:24]['VALUE'], c='blue', label='1949')
plt.plot(df[24:36]['MONTH'], df[24:36]['VALUE'], c='green', label='1950')
plt.plot(df[36:48]['MONTH'], df[36:48]['VALUE'], c='orange', label='1951')
plt.plot(df[48:60]['MONTH'], df[48:60]['VALUE'], c='black', label='1952')
plt.legend()
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel('Months, Integer')
plt.ylabel('Unemployment Rate, Percent')
plt.show()
```
